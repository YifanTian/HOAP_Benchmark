```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
output_only_n1ql = pd.read_csv('./AQA_8nodes_n1ql_throughput.csv')
output_only_n1ql
```

```python
plt.figure(figsize=(14,12))
idx = 0
label_list = [0,1,2,3,4]
for a in [2,3,4,5,6]:
    n1ql_list = []
    for i in range(7):
#         print(output_only_n1ql.iloc[i][3])
        n1ql_list.append(output_only_n1ql.iloc[i][a])
        

        
    plt.plot([2**i for i in range(7)], n1ql_list,'-o',label='A=' + str(label_list[idx]))
    idx+=1

plt.ylabel('n1ql throughput n/s',fontsize=24)
plt.xlabel('number of n1ql threads',fontsize=24)
plt.tick_params(labelsize=12)
plt.legend(fontsize = 16)
plt.savefig('figures/AQA_n1ql_throughput_vsn.png')
```

```python
connect_analytic_throughput = pd.read_csv('./AQA_8nodes_analytic_throughput.csv')
connect_analytic_throughput
```

```python
plt.figure(figsize=(14,12))
for i in range(len(connect_analytic_throughput)):
    if i == 0:
        n = 0
    else:
        n = 2**(i-1)
    plt.plot(connect_analytic_throughput.iloc[i][2:],'-o',label=str(n)+' n1ql')
    
plt.ylabel('analytic throughput n/min',fontsize=24)
plt.xlabel('number of analytic threads',fontsize=24)
plt.legend(fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('figures/AQA_analytic_throughput_vsa.png')
```
